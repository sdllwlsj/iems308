```python
corpus = list(itertools.chain(*corpus))
```

```python
with open('all/percentage.csv','r', encoding = 'latin-1') as f:
    percent = []
    rows = f.readlines()
    for row in rows:
        row = row.rstrip()
        percent.append(row)
```

```python
import functools
```

```python
def combine_string(article):
    
    return (functools.reduce(lambda x,y: x + " " + y, article,''))
```

```python
news = map(lambda x: combine_string(x), corpus)
news = list(news)
```

```python
def get_num_percentage(article):
    
#     reg = '\d+(?:\.\d+)?(?:%|percent(?:age points)?)'
#     pattern = re.compile(reg)
    
    percents = re.findall(r'\d+(?:\.\d+)?(?:%|percent(?:age points)?)', article)
    percents = filter(None,percents)
    
    percents = list(percents)
    
    return percents
```

```python
def get_wo_percentage(article):
    

    
    percents = re.findall(r'\w+(?:\-\w+)? (?:percent(?:age points)?)',article)
    percents = filter(None, percents)
    
    percents = list(percents)
    
    return percents    
```

```python
my_percents = []
for i, article in enumerate(news):
    num_percents = get_num_percentage(article)
    
    wo_percents = get_wo_percentage(article)
    
    my_percents.extend(num_percents)
    my_percents.extend(wo_percents)
    
    if i % 500 == 0:
        print(f"The program has already processed with {i} articles")
    
```

```python
with open('found_percent.txt', 'w') as f:
    for percent in my_percents:
        f.write(percent,)
```
